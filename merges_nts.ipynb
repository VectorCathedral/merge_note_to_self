```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 5, 9],
    "Col_1": [1, 2, 3, 4, 5],
    "Col_2": [6, 7, 8, 9, 10],
    "Col_3": [11, 12, 13, 14, 15],
    "Col_4": ["apple", "orange", "banana", "strawberry", "raspberry"]
})

df1
```

```python
df2 = pd.DataFrame({
    "ID": [1, 1, 3, 5],
    "Col_A": [8, 9, 10, 11],
    "Col_B": [12, 13, 15, 17],
    "Col_4": ["apple", "orange", "banana", "kiwi"]
})

df2
```

```python
pd.merge(df1,df2,left_on="Col_2",right_on="Col_B")
```

```python
inner =pd.merge(df1,df2)
inner
#merging on multiple columns by default
```

### inner merge first looks for common columns(if you dont specify your left_on/right_on),in this case ,ID and Col_4 are common <br>
### then,these columns get used as keys <br>
### what it then does ,look through all records/rows in each dataframe ,if the value in ID:df1 matches ID:df2 <b>AND</b> Col_4:df1 matches Col_4:df2 <br>
### it keeps those records/rows and merges them together (big keyword here is <b>AND</b>)


# inner merges work something like AND gate logic ,where both records need to match to be added(ALL conditions must be true)

# The positiong of where <b>df1 and df2 </b> are ,are important when dealing with other joins  </br>
## lets first understand what left and right are 

```python
left=pd.merge(df1,df2,how="left")
left
```

### left join is ,taking all the records from the table on the left which in this case is df1 ,then matching those with the records on the right which is df2<br>
### so we keep everything from df1,and only records from df2 that match the ones from df1 ,and if they dont match we get NAN<br>
### if df1 and df2 were swapped youd get different results,meaning df2 is our left table and df1 is our right table

```python
right=pd.merge(df1,df2,how="right")
right
```

### right join bascially just invereses our left join <br>
### we are keeping everything from df2 and the matching records from df1 ,non matching columns == NAN

```python
outer=pd.merge(df1,df2,how="outer")
outer
```

# Outer join just combines everything together ,whether there are matched records or not ,columns that dont match == NAN

### One thing to take note of is ,if you specfiy the columns to merge on with an inner ,if those columns dont have any matching records ,youll get nothing <br>
## Remeber the <b>AND</b> gate logic when working with inner merge/join

```python
nothing_merge=pd.merge(df1,df2,left_on="Col_1",right_on="Col_B",how="inner")
nothing_merge
```
